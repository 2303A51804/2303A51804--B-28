```python
import numpy as np
import matplotlib.pyplot as plt
monthly_temps_celsius = np.array([3.2, 4.8, 9.5, 14.0, 18.2, 22.0, 25.4, 24.8, 20.1, 14.5, 8.2, 4.0])

months = np.array([
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])
import numpy as np
import matplotlib.pyplot as plt


monthly_temps_celsius = np.array([3.2, 4.8, 9.5, 14.0, 18.2, 22.0, 25.4, 24.8, 20.1, 14.5, 8.2, 4.0])

months = np.array([
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])

avg_temp = np.mean(monthly_temps_celsius)
print(f"Average Temperature: {avg_temp:.2f}°C")


hottest_index = np.argmax(monthly_temps_celsius)
coldest_index = np.argmin(monthly_temps_celsius)
print(f"Hottest Month: {months[hottest_index]} ({monthly_temps_celsius[hottest_index]}°C)")
print(f"Coldest Month: {months[coldest_index]} ({monthly_temps_celsius[coldest_index]}°C)")


normalized_temps = (monthly_temps_celsius - monthly_temps_celsius.min()) / (monthly_temps_celsius.max() - monthly_temps_celsius.min())
print("Normalized Temperatures:")
print(np.round(normalized_temps, 2))

above_avg_indices = np.where(monthly_temps_celsius > avg_temp)[0]
print("Months Above Average Temperature:")
for i in above_avg_indices:
    print(f"{months[i]} ({monthly_temps_celsius[i]}°C)")


monthly_temps_fahrenheit = monthly_temps_celsius * 9/5 + 32
print("Temperatures in Fahrenheit:")
print(np.round(monthly_temps_fahrenheit, 1))


plt.figure(figsize=(10, 6))
plt.plot(months, monthly_temps_celsius, marker='o', label='Temperature (°C)', color='blue')
plt.axhline(y=avg_temp, color='red', linestyle='--', label=f'Avg Temp ({avg_temp:.1f}°C)')
plt.title("Monthly Temperature Trend")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd


data = {
    'Order ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', 'Phone', 'Monitor', 'Tablet', 'Monitor'],
    'Quantity': [1, 3, 2, 2, 5, 1, 4, 2],
    'Price': [1200, 400, 300, 1100, 420, 800, 310, 850]
}

df = pd.DataFrame(data)


df['Total Sale'] = df['Quantity'] * df['Price']

print(" Original Data with Total Sale:")
print(df)


high_value_orders = df[df['Total Sale'] > 1000]
print("\n Orders with Total Sale > 1000:")
print(high_value_orders)

product_sales = df.groupby('Product')['Quantity'].sum().reset_index()
most_sold_product = product_sales.sort_values(by='Quantity', ascending=False).iloc[0]

print(f"\n Most Sold Product: {most_sold_product['Product']} (Quantity: {most_sold_product['Quantity']})")



def calculate_discount(sale):
    if sale > 2000:
        return 0.15
    elif sale > 1000:
        return 0.10
    else:
        return 0.05

df['Discount Rate'] = df['Total Sale'].apply(calculate_discount)
df['Discount Amount'] = df['Total Sale'] * df['Discount Rate']
df['Final Sale'] = df['Total Sale'] - df['Discount Amount']

print("\n Data with Discount Columns:")
print(df)


output_file = 'ecommerce_sales_summary.xlsx'
df.to_excel(output_file, index=False)
print(f"\n Summary exported to {output_file}")
```
